```python
from scipy.stats import kendalltau

# Perform Mann-Kendall Trend Test

# Calculate the percentage of True values for each year
proportion_true = trendid2.groupby(['year'])['proovis_vähemalt_uks_ei_vasta_nõuetele'].apply(lambda x: (x == True).sum() / x.count() * 100).reset_index(name='percentage_true')

tau, p_value = kendalltau(proportion_true["year"], proportion_true["percentage_true"])
print(f'Mann-Kendall Tau: {tau}')
print(f'P-value: {p_value}')
if p_value < 0.05:
    trend = 'decreasing' if tau < 0 else 'increasing'
    print(f'The trend is statistically significant and {trend}.')
else:
    print('The trend is not statistically significant.')
print('---')
```

```python
result_corr = result_norm[["year","maakond","proportion_of_failed_tests","Bioloogiline hapnikutarve (BHT7)"]]
```

```python
from scipy import stats

# Define the significance level
alpha = 0.05

# Group the data by 'maakond' and perform the calculations for each group
results = {}
for maakond, group in result_corr.groupby('maakond'):
    # Calculate Pearson correlation coefficient
    correlation, _ = stats.pearsonr(group['proportion_of_failed_tests'], group['Bioloogiline hapnikutarve (BHT7)'])
    
    # Perform t-test for correlation
    t_score = correlation * (len(group) - 2)**0.5 / (1 - correlation**2)**0.5
    p_value = stats.t.sf(abs(t_score), df=len(group)-2)*2  # two-sided p-value
    
    # Decision to reject the null hypothesis
    reject_null = p_value < alpha
    
    # Store the results
    results[maakond] = {
        'Pearson correlation coefficient': correlation,
        't-score': t_score,
        'p-value': p_value,
        'Reject Null Hypothesis': reject_null
    }

# Output the results
for maakond, res in results.items():
    print(f'Maakond: {maakond}')
    print(f'Pearson correlation coefficient: {res["Pearson correlation coefficient"]}')
    print(f't-score: {res["t-score"]}, p-value: {res["p-value"]}')
    print(f'Reject Null Hypothesis: {"Yes" if res["Reject Null Hypothesis"] else "No"}')
    print('---')
```
